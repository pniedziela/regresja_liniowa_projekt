### Regresja liniowa

Regresja liniowa jest jedyn z najprostszych przykładów uczenia nadzorowanego. Pozwala ona na przewidywanie związku pomiędzy analizowanymi danami a także na wytyczenie pewnych linii trendu. Znalduje szerokie zastosowanie w szacowaniu wartości rzeczywistych, takich jak: 
- Ceny domów/mieszkań (jak w przypadku obecnej pracy)  
- Sprzedaży na jedną osobę  
- Analizy przychodów firm  


Regresja liniowa wykonuje pewne zadanie, polegające na predykcji wartości zależnej (y) bazując na pewnej wartości niezależnej (x). W związku z tym, opisywana technika "znajduje" relacje liniową pomiędzy wartością (x) oraz (y).  
Funkcją opisującą taką relację będzie: 


<center>$y = θ_1 + θ_2x$</center>

Gdzie podczas trenowania modelu:  

y – zmienna objaśniana,  
x – zmienna objaśniająca,  
$θ_1$ – współczynnik kierunkowy regresji (ang. *slope*),  
$θ_2$ – wyraz wolny (ang. *intercept*).  

W czasie trenowania modelu dobierana jest najlepsza linia do predykcji wartości *y* przy danej wartości *x*. Odbywa się to poprzez dobór najlepszych wartości dla: $θ_1$ oraz $θ_2$ i wykorzystywana do tego jest funkcja kosztu:

<center>$J = \frac{1}{n}  \sum \limits _{n=1} ^{n} (pred_i - y_i)^2$

Dla regresji liniowej funkcją kosztu jest RMSE (ang. *Root Mean Squared Error*), i do jej zadania należy obliczenie błędu pomiędzy wartością predykowaną a wartością prawdziwą (oczekiwaną). 
  
Dodatkowo aby model mógł się w prawidłowy sposób "uczyć" potrzebny jest sposób na uaktualnienie wartości  $θ_1$ oraz $θ_2$, w tym celu stosuje się motodę gradientu prostego. 

Jest to metoda optymalizacji służąca do znalezienia minimum lokalnego w zadaniej funkcji celu. Wartości początkowe dla  $θ_1$ oraz $θ_2$ generowane są losowo, aby w kolejnych iteracjach aktualizować je aż do uzyskania minimalnej wartości kosztu. 

### Regresja liniowa w predykcji cen mieszkań

W związku z możliwościami regresji liniowej w postaci przewidywania zwiazków pomiędzy analizowanymi danymi czy predykcji pewnych danych zależnych, możliwe jest wykorzystanie jej na rynku nieruchomości. Może znaleźć zastosowanie w predykcji oczekiwaniej ceny mieszkań, bądź analizy atrakcyjności poszczególnych obszarów miejskich ze względów na różne parametry (gęstość zabudowy, sąsiedztwo obszarów przemysłowych, poziom zieleni, odległość od głównych arterii miejskich). Z drugiej natomiast strony umożliwia analizę najbardziej atrakcyjnych oraz pożądanych (a zarazem wpływających na cenę) cech (ma przykład prywatne miejsce parkingowe). 

### Opis uzyskanych wyników oraz ocena modelu

Model uzyskał wyniki na poziomie:  
RMSE: 0.1272489907926278  

Wynik RMSE jest zależny od kontekstu, w tym wypadku przy średniej wartości ("cenie") na poziomie 12.06 (dane poddane zostały transformacji logarytmicznej) i RMSE na poziomie 0.1272, przy teoretycznej cenie domu w wysokości 100 tys. dolarów, błąd wyniósłby ~1010 dolarów. Wynik taki, w kontekście predykcji cen mieszkań uznać można za akceptowalny.
